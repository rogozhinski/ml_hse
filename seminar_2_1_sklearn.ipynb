```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import warnings
warnings.simplefilter("ignore")
%matplotlib inline
sns.set(style="darkgrid")
```

```python
np.random.seed(36)
x_func = np.linspace(0, 1, 100)
y_func = np.cos(1.5 * np.pi * x_func)
```

```python
x_obj = np.random.uniform(0, 1, 30)
y_obj = np.cos(1.5 * np.pi * x_obj) + np.random.normal(0, 0.1, x_obj.shape)
```

```python
x_regr = x_obj[:,None]
y_regr = y_obj[:,None]
```

```python
from sklearn.preprocessing import PolynomialFeatures
polynom = PolynomialFeatures(degree=20, include_bias=False)
x_learn = polynom.fit_transform(x_regr)
x_learn
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=0.001)
ridge.fit(x_learn, y_regr)
```

```python
x_pred = polynom.fit_transform(x_func[:,None])
y_pred = ridge.predict(x_pred)
```
